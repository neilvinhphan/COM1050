# W6A4

```python
lst = list(map(str, input().split()))
result = {}

for i in range(0, len(lst), 2):
    if not lst[i] in result:
        result.setdefault(lst[i], [])
        result[lst[i]].append(lst[i+1])
    else:
        result[lst[i]].append(lst[i+1])

print(result)
    
```

# W6A5

```python
lst = list(map(int, input().split()))
result = {
    'positives':[],
    'negatives':[],
    'zeros':[]
}

for i in lst:
    if i == 0:
        result['zeros'].append(i)
    elif i > 0:
        result['positives'].append(i)
    else:
        result['negatives'].append(i)

print(result)
```

# W6A9

```python
in1 = list(map(str, input().split()))
in2 = list(map(str, input().split()))

lst1 = {x: y for x, y in zip(in1[0::2],in1[1::2])}
lst2 = {x: y for x, y in zip(in2[0::2],in2[1::2])}

result = {}

for i in lst1:
    if i not in lst2:
        result.update({i:lst1.get(i)})
    else:
        result.update({i:(lst1.get(i) + lst2.get(i))})
        lst2.pop(i)
    
result.update(lst2)

print(dict(sorted(result.items())))
```

# W6A10

```python
lst = list(map(int, input().split()))
k = int(input())
result = []

for i in range(len(lst)):
    for s in range(i, len(lst)):
        if lst[i] + lst[s] == k:
            result.append((lst[i], lst[s]))

result.sort()
print(result)
```
