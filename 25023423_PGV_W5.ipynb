```python
def Max_Of_Two(x, y):
    maxn = x
    if x < y:
        maxn = y
    return maxn

a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
def snt_check(x):
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return False
    return True

snt_check(int(input()))
```

# W5A10

```python
a, b = map(str, input().split())
strs = {}
alpb = ["a", "b", "c", "d", "e", "f", "g", "h", "i"," j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z" ]

def mapping_check(x, y):
    count = 0
    for i in x:
        if i == y[count]: 
            count += 1
            continue
        else:
            if not strs.get(i) and y[count] in alpb:
                strs.update({i:y[count]})
                alpb.remove(y[count])
            elif strs.get(i) != y[count]:
                return False
        count += 1
    return True

print(mapping_check(a, b))
```
