```python
def snt_check(x):
    if x == 1:
        return False
    elif x == 2:
        return True
    else:
        for s in range(2, (x//2)+1):
            if x == s:
                continue
            elif x % s == 0:
                return False

    return True

a, b = map(int, input().split())
result = 0

for i in range(a, b+1):
    if snt_check(i):
        result += i

print(result)
```

# W4A14

```python
m, n = map(int, input().split())
raw = []

def find(x):
    for i in range(1, x+1):
        if  m % i == 0 and n % i == 0:
            raw.append(i)

def max(lst):
    result = lst[0]
    for s in lst:
        if s > result:
            result = s
    return result

if m > n:
    find(m)
    print(max(raw))
elif m < n:
    find(n)
    print(max(raw))
```
